```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib inline
```

```
#leitura dos dados
df = pd.read_csv('https://raw.githubusercontent.com/CristinaPineda/desafio_serasa/master/datasetMaria.csv')
df
```

```
#setando o ndice
#df = df.set_index('App')
#df
```

```
#estatsticas bsicas
df.describe()
```

```
#adicionando coluna de classificao de estrelas de acordo com os percentis do describe 
df1 = pd.DataFrame({'estrelas': list(df.loc[:,'estrelas'])})
conditions = [
              (df['estrelas'] > 4.65),
              (df['estrelas'] >= 4.50)
              ]
choices = ['otimo', 'bom']
df['classificacao'] = np.select(conditions, choices, default= 'pode melhorar')
df
```

Considerando uma amostra mnima de concorrentes que possuem caracteristicas que de acordo (ou caracteristicas similares e/ou prximas) com a nossa breve pesquisa podemos ter uma geso geral de como anda as avalies dos apps de finanas gratuitos disponivies no mercado hoje.

```
# classificao geral dos apps
df['classificacao'].value_counts().plot.bar(figsize=(10,5),
                                            rot= 0,
                                            title= '# Classificao por estrelas');
```

```
# classificao por tamanho do aplicativofig, ax = plt.subplots()
plt.figure(figsize = (10, 10))
plt.barh(df['App'], df['tamanho'])

plt.xlabel("Aplicativos")
plt.ylabel("Tamanho em megas")

plt.show()
```

```
#Aplicativos segundo a classificao das estrelas
plt.figure(figsize = (17, 5))
plt.bar(df['App'], df['classificacao'])
plt.title('Aplicativos segundo a classificao das estrelas')
plt.xlabel("Aplicativos")
plt.ylabel("Classificao")

plt.show()
```
